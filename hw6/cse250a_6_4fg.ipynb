```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Qprime(y):
    return y - np.tanh(y)
    
def f(x):
    return np.log(np.cosh(x))

def fprime(x):
    return np.tanh(x)
    
def fprimeprime(x):
    return 1/(np.cosh(x))**2

def newtonsmethod(x):
    arr = []
    curval = f(x)
    arr.append(x)
    print("f(x) at x={}: {}".format(x, curval))
    x = x - (fprime(x)/fprimeprime(x))
    newval = f(x)
    arr.append(x)
    print("f(x) at x={}: {}".format(x, curval))
    while newval != curval:
        curval = newval
        x = x - (fprime(x)/fprimeprime(x))
        newval = f(x)
        arr.append(x)
        print("f(x) at x={}: {}".format(x, curval))
        
    return arr

def findminima(x):
    arr = []
    curval = f(x)
    arr.append(x)
    print("f(x) at x={}: {}".format(x, curval))
    x = Qprime(x)
    newval = f(x)
    arr.append(x)
    print("f(x) at x={}: {}".format(x, curval))
    while newval != curval:
        curval = newval
        x = Qprime(x)
        newval = f(x)
        arr.append(x)
        print("f(x) at x={}: {}".format(x, curval))
        
    return arr
```

### 6.4 f) 

```python
arr = findminima(-2)
plt.plot(arr, 'go-')
plt.xlabel("Iterations")
plt.ylabel("Xn")
plt.show()
```

```python
arr = findminima(3)
plt.plot(arr, 'go-')
plt.xlabel("Iterations")
plt.ylabel("Xn")
plt.show()
```

### 6.4g) Newton's method does not converge when we set x0= -2 or x0 = 3. This is because in one of the iterations the double derivative is becoming 0. When we start with x0=-2 or 3 the subsequent xn values get even larger and we end up at the asymptotic maxima at f(x) = infinity.

```python
arr = newtonsmethod(-2)
plt.plot(arr, 'go-')
plt.xlabel("Iterations")
plt.ylabel("Xn")
plt.show()
```

```python
arr = newtonsmethod(3)
plt.plot(arr, 'go-')
plt.xlabel("Iterations")
plt.ylabel("Xn")
plt.show()
```
